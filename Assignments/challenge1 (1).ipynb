 Challenge: Student Marks Analyzer

Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names


```python

class StudentMarkAnalyzer:

  def __init__(self):
    self.students = {}

  def add_students(self):
    while True:
      name = input("Enter student name(or type 'done' to finish):").strip()

      if name.lower() == "done":
        break

      if name in self.students:
        print("Student already exist!! Try different name\n")
        continue

      try:
        marks = int(input(f"Enter marks for {name}:"))
        self.students[name] = marks
      except ValueError:
        print("Invalid input! Marks must be int value.")

  def Generate_Report(self):

    if not self.students:
      print("No student data avilable!!")

    marks_list = list(self.students.values())
    total_students = len(self.students)
    avg_makrs = sum(marks_list) / total_students

    highest = max(marks_list)
    lowest = min(marks_list)

    top_students = [name for name, mark in self.students.items() if mark == max(marks_list)]
    low_students = [name for name, mark in self.students.items() if mark == min(marks_list)]

    #Report Output
    print("\n"+"=" * 40)
    print("STUDENT MARK REPORT")
    print("=" * 40)

    print(f"Total Students:{total_students}")
    print(f"Avg Makrs:{avg_makrs:2f}")

    print(f"\n🏆 Highest Marks:{highest}")
    print(f"Top students(s):" + ",".join(top_students))

    print(f"\n📉 lowest Marks:{lowest}")
    print(f"Low students(s):" + ",".join(low_students))

    print("=" * 40)

#Main program
analyzer = StudentMarkAnalyzer()
analyzer.add_students()
analyzer.Generate_Report()
```
