```python
# 6. Write a Python program to create a class representing a stack data structure. Include methods for pushing
# and popping elements.
class Stack:
    def __init__(self):
        self.items = []

    def push(self, *item):
        self.items.extend(item)  #to push one item at a time  use append
        return f"{item} pushed to stack"

    def pop(self):
        if not self.items:
            return "Stack is empty"
        return self.items.pop()

    def peek(self):
        if not self.items:
            return "Stack is empty"
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def display(self):
        return self.items

s = Stack()

print(s.push(10))
print(s.push(20,30,40))

print("Stack:", s.display())
print("Popped:", s.pop())
print("Top element:", s.peek())
print("Is empty:", s.is_empty())
```

```python
# 8. Write a Python program to create a class representing a shopping cart. Include methods for adding and
#  removing items, and calculating the total price.

class ShoppingCart:

  def __init__(self):
    self.items = {}
    pass

  def adding(self, name, price, quantity=1):
    if name in self.items:
      self.items[name] = (price, self.items[1] + quantity)
    else:
      self.items[name] = (price, quantity)

    return f"Item Added: {name} X {quantity} successfully."

  def removing(self,name, quantity):

    if name in self.items:
      current_quantity = self.items[name][1]
      if quantity > current_quantity:
        del self.items[name]

      else:
        self.items[name] = (self.items[name][0], current_quantity - quantity)

    return f"Item removed: {name} X {quantity} successfully."

  def display_items(self):
    if not self.items:
      return "Cart is empty."
    else:
      result =""
      for name, (price, quantity)  in self.items.items():
        result += f"{name} : {price} * {quantity} = {price * quantity} Rs\n"
      return result


  def Total(self):
    total = 0
    for price,quantity in self.items.values():
      total += (price * quantity)
    return f"Total sum ={total}"

s1 = ShoppingCart()
print(s1.adding("apple", 50, 1))
print(s1.adding("mango", 50, 1))
print(s1.Total())
print(s1.display_items())
```

```python
# 10. Write a Python program to create a class representing a queue data structure. Include methods for
# enqueueing and dequeueing elements.

class Queue:

  def __init__(self):
     self.items = []

  def enqueue(self, item):
    self.items.append(item)
    return f"{item} added to queue"

  def dequeue(self):
    if not self.items:
      return "Queue is empty"
    return self.items.pop(0)

  def display(self):
    return self.items

q = Queue()

print(q.enqueue(10))
print(q.enqueue(20))
print(q.enqueue(30))

print("Queue:", q.display())
print("Dequeued:", q.dequeue())
```

```python
# 14. Write a program that uses datetime module within a class. Enter manufacturing date and expiry date of the
#  product. The program must display the years, months and days that are left for expiry.
from datetime import datetime

class Product:

    def __init__(self):
        self.__mfg_date = None
        self.__exp_date = None

    def set_mfg_date(self, mfg):
        self.__mfg_date = datetime.strptime(mfg, "%d-%m-%Y")

    def set_exp_date(self, exp):
        self.__exp_date = datetime.strptime(exp, "%d-%m-%Y")

    def time_to_expiry(self):
        today = datetime.now()

        if self.__exp_date < today:
            print("Product already expired!!!")
            return

        diff = self.__exp_date - today

        days = diff.days
        years = days // 365
        months = (days % 365) // 30
        remaining_days = (days % 365) % 30

        print(f"Time left for expiry: {years} years, {months} months, {remaining_days} days")


p = Product()

mfg = input("Enter manufacturing date (DD-MM-YYYY): ")
exp = input("Enter expiry date (DD-MM-YYYY): ")

p.set_mfg_date(mfg)
p.set_exp_date(exp)

p.time_to_expiry()
```

```python
# 19. Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and methods like add_item_to_menu, book_tables, and customer_order.

# 	Perform the following tasks now:

# 	Now add items to the menu.
# 	Make table reservations.
# 	Take customer orders.
# 	Print the menu.
# 	Print table reservations.
# 	Print customer orders.
# 	Note: Use dictionaries and lists to store the data.

class Restaurant:

    def __init__(self):
        self.menu_items = {}          # {item: price}
        self.book_table = []          # list of table numbers
        self.customer_orders = []     # list of orders

    # Add item to menu
    def add_item_to_menu(self, item, price):
        self.menu_items[item] = price

    # Book table
    def book_tables(self, table_no):
        self.book_table.append(table_no)

    # Take customer order
    def customer_order(self, table_no, item):
        if item in self.menu_items:
            self.customer_orders.append({"table": table_no, "item": item})
        else:
            print(f"{item} is not available in menu!")

    # Print menu
    def print_menu(self):
        print("\n--- MENU ---")
        for item, price in self.menu_items.items():
            print(f"{item:<15} : ₹{price:>9}")

    # Print table reservations
    def print_tables(self):
        print("\n--- TABLE RESERVATIONS ---")
        print(self.book_table)

    # Print customer orders
    def print_orders(self):
        print("\n--- CUSTOMER ORDERS ---")
        for order in self.customer_orders:
            print(f"Table {order['table']} ordered {order['item']}")

# ------------------ TEST ------------------

r = Restaurant()

# Add items
r.add_item_to_menu("Pizza", 250)
r.add_item_to_menu("Burger", 120)
r.add_item_to_menu("Pasta", 200)

# Book tables
r.book_tables(1)
r.book_tables(2)

# Orders
r.customer_order(1, "Pizza")
r.customer_order(2, "Burger")
r.customer_order(1, "Pasta")

# Display
r.print_menu()
r.print_tables()
r.print_orders()
```

```python
# 20. Write a Python class BankAccount with attributes like account_number, balance, date_of_opening and
# customer_name, and methods like deposit, withdraw, and check_balance.

from datetime import datetime

class BankAccount:

    def __init__(self, account_number, customer_name, balance):
        self.account_number = account_number
        self.customer_name = customer_name
        self.balance = balance
        self.date_of_opening = datetime.now().strftime("%d-%m-%Y")

    # Deposit method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully.")
        else:
            print("Invalid deposit amount!")

    # Withdraw method
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        elif amount <= 0:
            print("Invalid withdrawal amount!")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully.")

    # Check balance
    def check_balance(self):
        print(f"Current Balance: ₹{self.balance}")

    # Display account details
    def display_details(self):
        print("\n--- ACCOUNT DETAILS ---")
        print("Account Number :", self.account_number)
        print("Customer Name  :", self.customer_name)
        print("Opening Date   :", self.date_of_opening)
        print("Balance        : ₹", self.balance)


# ------------------ TEST ------------------

acc = BankAccount("123456789", "Vicky", 10000)

acc.display_details()

acc.deposit(2000)
acc.withdraw(3000)
acc.check_balance()
```

```python
# 21. Write a Python class Inventory with attributes like item_id, item_name, stock_count, and price, and methods
# like add_item, update_item, and check_item_details.

class Inventory:

    def __init__(self):
        # store items as dictionary
        # item_id : {name, stock, price}
        self.items = {}

    # Add item
    def add_item(self, item_id, item_name, stock_count, price):
        self.items[item_id] = {
            "name": item_name,
            "stock": stock_count,
            "price": price
        }
        print(f"Item '{item_name}' added successfully!")

    # Update item
    def update_item(self, item_id, stock_count=None, price=None):
        if item_id in self.items:
            if stock_count is not None:
                self.items[item_id]["stock"] = stock_count
            if price is not None:
                self.items[item_id]["price"] = price
            print("Item updated successfully!")
        else:
            print("Item not found!")

    # Check item details
    def check_item_details(self, item_id):
        if item_id in self.items:
            item = self.items[item_id]
            print("\n--- ITEM DETAILS ---")
            print("ID    :", item_id)
            print("Name  :", item["name"])
            print("Stock :", item["stock"])
            print("Price : ₹", item["price"])
        else:
            print("Item not found!")

# ------------------ TEST ------------------

inv = Inventory()

# Add items
inv.add_item(101, "Laptop", 10, 50000)
inv.add_item(102, "Mouse", 50, 500)

# Update item
inv.update_item(101, stock_count=8, price=48000)

# Check details
inv.check_item_details(101)
inv.check_item_details(102)
```
