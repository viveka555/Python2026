```python
# 2. Write a Python program to read first n lines of a file.

n = int(input("Enter no of lines: "))
with open("file1.txt", "r") as file:
  lines = file.readlines()  #gives list of lines
  print("".join(lines[:n])) #print first n lines.
```

```python
# 7. Write a python program to find the longest words.
with open("file1.txt", "r") as file:
  words = file.read().split()

  max_len = max(len(word) for word in words)
  longest_word = [word for word in words if len(word) == max_len ]
print(longest_word)
```

```python
# 9. Write a Python program to count the frequency of words in a file.
with open("file1.txt", "r") as file:
  words = file.read().lower().split()
  word_freq = {}
  for word in words:
    word_freq[word] = word_freq.get(word, 0) + 1
print(word_freq)
```

```python
# 11. Write a Python program to write a list to a file.
list1 = ["vivek", "rohit", "sachin", "kohli"]
with open("file1.txt", "a") as file:
  for item in list1:
    file.write(item + "\n")
```

```python
# 12. Write a Python program to copy the contents of a file to another file .
with open("file1.txt", "r") as source, open("file.txt", "w") as destination:
  for line in source:
    destination.write(line)
```

```python
# 13. Write a Python program to read a random line from a file.
import random
with open("file.txt", "r") as file:
  lines = file.readlines()
random_lines = random.choice(lines)

print(random_lines)
```
