```python
# 4. Write a Python function to reverse a string.
Str1 = "1234abcd"

def reverse_str(n):
  reverse = ""
  for i in n:
    reverse = i + reverse
  return reverse

print(reverse_str(Str1))


# 	Expected Output : "dcba4321"
```

```python
# 5. Write a Python function to check whether a number falls within a given range.
def is_in_range(n,start,stop):
  if start <= n <= stop:
    return True
  else:
    return False
print(is_in_range(5,2,10))
```

```python
# 6. Write a Python function that accepts a string and counts the number of upper and lower case letters.
str1 = 'The quick Brow Fox'

def count_lower_upper(s):
  lowers = 0
  uppers = 0
  for char in s:
    if char.islower():
      lowers += 1
    elif char.isupper():
      uppers += 1
  return f" No. of Lower case characters :{lowers}\n No. of Upper case Characters :{uppers}"

print(count_lower_upper(str1))
# 	Expected Output :
# 	No. of Upper case characters : 3
# 	No. of Lower case Characters : 12
```

```python
# 8. Write a Python function that checks whether a passed string is a palindrome or not.
str1 = "madam"

def is_palindrome(s):
  return s == s[::-1]

print(is_palindrome(str1))

# 	Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.
```

```python
# 22. Write a Python program to check whether a given string contains a capital letter, a lower case letter,
# a number and a minimum length using lambda.

check_string = lambda s: (
    any(c.isupper() for c in s) and
    any(c.islower() for c in s) and
    any(c.isdigit() for c in s) and
    len(s) >= 8
)

print(check_string("Vivek123"))   # True
print(check_string("vivek123"))   # False (no uppercase)
print(check_string("VIVEK123"))   # False (no lowercase)
print(check_string("Vivek"))      # False (too short, no digit)
```
