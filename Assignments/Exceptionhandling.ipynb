```python
# 1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.
try:
  num = int(input("Enter your number: "))
  print(10/num)
except ZeroDivisionError:
  print("You can not divide by zero!!!")
```

```python

# 2. Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input
# is not a valid integer.
try:
  num = input("Enter your number: ")

  if not num.isdigit():
    raise ValueError("It's not a valid integer!!!")

  num = int(num)
  print(f"your input:{num}")
except ValueError as e:
  print(f"Error:{e}")

```

```python
# 3. Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist.
try:
  with open("file.txt" "r") as file:
    content = file.read()
    print(content)
except FileNotFoundError:
  print("file does not exist!!")
```

```python
# 4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception
#  if the inputs are not numerical.
try:
  num1 = input("Enter your number: ")
  num2 = input("Enter your number: ")
  if not num1.isnumeric() and not num2.isnumeric():
    raise TypeError("input is not a numeric!!")
  num3 = float(num1)
  num4 = float(num2)
  print(f"Your numbers are:{num3},{num4}")
except TypeError as e:
  print(f"Error:{e}")
```

```python
# 5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.

try:
  with open("file.txt", "r") as file:
    print(file.read())
except PermissionError:
  print("You dont have permission to open this file")
except FileNotFoundError:
  print("file not found!!!")
```

```python
# 6. Write a Python program that executes an operation on a list and handles an IndexError exception if the index
# is out of range.
try:
  lst = [1,2,3,4,5]
  inp = int(input("Enter index: "))
  print("your input",lst[inp])
except IndexError:
  print("input out of index")
except ValueError:
  print("Error: please enter valid int")
```

```python
# 7. Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic
#  error.
try:
  a = int(input("Enter your number:"))
  print(10/a)
except ArithmeticError:
  print("mathematical Error Occured!!")
except ValueError:
  print("please Enter valid input")
```
