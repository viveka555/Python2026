Hospital Patient Queue Management System
The Hospital Patient Queue Management System is a Python-based command-line application designed to manage patient queues in a hospital. It utilizes object-oriented programming (OOP) principles and includes three main classes:

**Patient**

**Specialization**

**OperationsManager **

This project provides a simple yet effective way to handle patient data and queue management within a hospital setting.

The Hospital Patient Queue Management System is an illustration of OOP concepts in Python. It consists of three primary classes, each serving a specific purpose:

**Patient:**

The Patient class represents an individual patient and includes the following attributes:

•	name: The name of the patient.

•	status: The patient's status, which can be 0 (normal), 1 (urgent), or 2 (super-urgent).

This class provides methods for string representation and status formatting for patients.

**Specialization:**

The Specialization class manages patient queues within different specializations. It offers functionalities such as:

•	Adding patients with various urgency levels.

•	Retrieving the next patient from the queue.

•	Removing patients by name.

•	Checking queue capacity.

**OperationsManager:**

The OperationsManager class serves as the user interface for interacting with the Specialization instances. Users can perform actions like:

•	Adding new patients to specializations.

•	Listing patients in specializations.

•	Retrieving the next patient.

•	Removing patients.

•	Ending the program gracefully.




```python
class Patient:
    def __init__(self, name, age, status):
        self.name = name
        self.age = age
        self.status = status

    def get_status_display(self):
        if self.status == 0:
            return "Normal"
        elif self.status == 1:
            return "Urgent"
        elif self.status == 2:
            return "Super-Urgent"
        else:
            return "Unknown"

    def __str__(self):
        return f"Patient Name: {self.name}, Age: {self.age}, Status: {self.get_status_display()}"


class Specialization:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.patients = {}

    # Add Patient
    def add_patient(self, name, age, status):
        key = name.lower()

        if key in self.patients:
            print("Patient already exists!\n")
            return

        if len(self.patients) >= self.capacity:
            print("Queue is full!!\n")
            return

        self.patients[key] = Patient(name, age, status)
        print("Patient added successfully\n")

    # List Patients
    def list_of_patients(self):
        if not self.patients:
            print("No patients in this specialization.\n")
            return

        print(f"\n--- {self.name} Patients ---")
        for p in self.patients.values():
            print(p)
        print()

    # Get Next Patient
    def get_next_patient(self):
        if not self.patients:
            print("No patient in Queue!\n")
            return

        patient = max(self.patients.values(), key=lambda p: p.status)
        print(f"Next Patient: {patient}\n")

    # Remove Patient
    def remove_patient(self, name):
        key = name.lower()

        if key not in self.patients:
            print("Patient not found!!\n")
            return

        del self.patients[key]
        print("Patient successfully removed!!\n")

    # Check Capacity
    def check_capacity(self):
        print(f"{len(self.patients)} / {self.capacity} patients\n")


class OperationsManager:
    def __init__(self):
        self.specializations = {
            1: Specialization("Cardiology", 5),
            2: Specialization("Neurology", 5),
            3: Specialization("Orthopedics", 5)
        }

    def run(self):
        while True:
            print("===== Hospital System =====")
            print("1. Add Patient")
            print("2. List Patients")
            print("3. Get Next Patient")
            print("4. Remove Patient")
            print("5. Check Capacity")
            print("6. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.add_patient()
            elif choice == "2":
                self.list_patients()
            elif choice == "3":
                self.get_next_patient()
            elif choice == "4":
                self.remove_patient()
            elif choice == "5":
                self.check_capacity()
            elif choice == "6":
                print("Exiting system...")
                break
            else:
                print("Invalid choice!\n")

    # ----------- Helper Methods -----------

    def get_specialization(self):
        print("\nSelect Specialization:")
        print("1. Cardiology")
        print("2. Neurology")
        print("3. Orthopedics")

        try:
            choice = int(input("Enter number: "))
            return self.specializations.get(choice)
        except:
            return None

    def add_patient(self):
        spec = self.get_specialization()
        if not spec:
            print("Invalid specialization!\n")
            return

        name = input("Enter name of patient: ")
        age = int(input("Enter age: "))
        status = int(input("Enter status (0=Normal, 1=Urgent, 2=Super Urgent): "))

        spec.add_patient(name, age, status)

    def list_patients(self):
        spec = self.get_specialization()
        if spec:
            spec.list_of_patients()
        else:
            print("Invalid specialization!!\n")

    def get_next_patient(self):
        spec = self.get_specialization()
        if spec:
            spec.get_next_patient()
        else:
            print("Invalid specialization!!\n")

    def remove_patient(self):
        spec = self.get_specialization()
        if not spec:
            print("Invalid specialization!!\n")
            return

        name = input("Enter patient name: ")
        spec.remove_patient(name)

    def check_capacity(self):
        spec = self.get_specialization()
        if spec:
            spec.check_capacity()
        else:
            print("Invalid specialization!!\n")


# ----------- Main Program -----------

system = OperationsManager()
system.run()
```
